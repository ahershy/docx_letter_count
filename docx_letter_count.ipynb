```python
#importing required packages
import docx2txt
import pandas as pd
import matplotlib.pyplot as plt

#converting docx file to text
text = docx2txt.process(r'C:\Users\Andrew\Desktop\Resumes\August_2022_Resume.docx')

#creating loop to count frequency of letters in text object
alphabet=['a','b','c','d','e','f','g','h','i','j','k','l','m','n','o','p','q','r','s','t','u','v','w','x','y','z']
frequency = []
for i in range(0,26):
    frequency.append(text.count(alphabet[i]))
    
#creating dataframe to build visualization from
total_log = pd.DataFrame({ 'letter': alphabet, 'frequency': frequency})
total_log

```

```python
#creating bar chart
plt.bar(total_log['letter'],total_log['frequency'])
plt.title('resume letter count')
plt.xlabel('letters')
plt.ylabel('count')
plt.show()
```
